```python
import pandas as pd
```

```python
df = pd.read_csv("HR_Analytics.csv")
df.tail()
```

```python
# Change the column name from 'sales' to 'job' because that column have same value(sales)
df=df.rename(columns={'sales':'job'})
df.tail()
```

```python
#colum 'job' & 'salary' encoding
from sklearn.preprocessing import LabelEncoder

le=LabelEncoder()
df['job']= le.fit_transform(df['job'])
df['salary']= le.fit_transform(df['salary'])
df.tail()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.pairplot(df)
plt.show()
```

```python
sns.barplot(x = 'time_spend_company', y = 'left', data = df)
sns.plt.title('Left over time spend at company (barplot)')
sns.factorplot(x= 'time_spend_company', y = 'left', data = df, size = 5)
sns.plt.title('Left over time spend at company (factorplot)')
plt.show()
```
